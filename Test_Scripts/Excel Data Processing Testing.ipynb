```python
def reorder_rename_cols(old_df):
    list_of_column_names = ['First Name',
                            'Last Name',
                            'Email Address',
                            'Health Screening (Worksite or Provider)',
                            'Last Health Assessment (VP Health Check)',
                            'Last Digital Coaching (1 VP Journey)',
                            'Activity Campaigns (VP Team Challenges) **',
                            'Initiatives Completed',
                            'Reimbursements Received',
                            'Reimbursement Left',
                            'Last Email Sent (Y/N)',
                            'Current Employee (Y/N)']

    old_df = old_df[list_of_column_names].copy()
    renamed_df = old_df.rename(columns={'Health Screening (Worksite or Provider)': 'Health Screening',
                                        'Last Health Assessment (VP Health Check)': 'Health Assessment',
                                        'Activity Campaigns (VP Team Challenges) **': 'Activity Campaigns'})
    return renamed_df
```

```python
def completed_check(person_data, item, output):

    if person_data[item] == 1:
        completed.append(output)
    else:
        un_completed.append(output)
```

```python
def data_gather(person_data):
    
    requirements = ({3: 'Health Screening',
                     4: 'Health Assessment',
                     5: 'Digital Coaching (VP Journey)',
                     6: 'First Activity Campaign',
                     7: 'Second Activity Campaign',
                 })
    
    f_name = person_data[0]
    l_name = person_data[1]
    email = person_data[2]
    completed = []
    un_completed = []
    
    for j in range(3,6):
        completed_check(person_data, j, requirements[j])
    
    return f_name, l_name, email, completed, un_completed
        
    
```
