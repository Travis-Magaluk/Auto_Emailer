```python
# Only Columns we need here sorting out the unnecessary column names. 

list_of_column_names = ['Last Name', 'First Name', 'Member Status', 'Plan Status', 'Last Health Assessment \
(VP Health Check)', 'Last Digital Coaching (1 VP Journey)', 'Health Screening (Worksite or Provider)','Activity \
Campaigns (VP Team Challenges) **']

df_sort = df1[list_of_column_names]
```

```python
df_sort.head()

# Filter out only subscribers.
df_sort = df_sort[df_sort['Member Status'] == 'SUBSCRIBER']

df_sort.head(10)
```

```python
# Replacing any blank cells with NaN
df_sort = df_sort.replace(r'^\s*$', np.nan, regex=True)
```

```python
df_sort.head(20)
```

```python
# Pulling columns where we want to see if people have done something or not.  These columns are being prepared 
# for converting to 0's or 1's as a yes or no response. 

df_parse = df_sort[['Last Health Assessment (VP Health Check)', 'Last Digital Coaching (1 VP Journey)', 'Health Screening (Worksite or Provider)']]
df_parse.head(20)


```

```python
# Dropping columns above from df_sort. Will remerge later. 

cols = ['Last Health Assessment (VP Health Check)', 'Last Digital Coaching (1 VP Journey)', 'Health Screening (Worksite or Provider)']
df_sort = df_sort.drop(labels=cols,axis=1)


```

```python
df_sort.head(20)
```

```python
df_sort.dtypes
```

```python
df_parse.head()
```

```python
# This is converting anything that is null to a zero "int" dtype and anything notnull to a one "int" dtype.
 
df_parse = df_parse.notnull().astype('int')
```

```python
df_parse.head()
```

```python
# Testing iloc function and getting a feel for how it will come out. 
# Feel that it will be easier for program to run line by line through the data rather than create lists for each
# column. 

s1 = df_sort.iloc[1]
```

```python
# Merging these two dfs back together to one cleaned up file from the health insurance company. 

result = pd.concat([df_sort, df_parse], axis=1)
```
